```python
from sklearn.datasets import load_iris
iris = load_iris()
X, y = iris.data, iris.target
X.shape, y.shape
```

```python
from sklearn.decomposition import PCA
pca = PCA()
Xpca = pca.fit_transform(X)
Xpca.shape
```

```python
import numpy as np
np.var(X, axis=0), np.var(Xpca, axis=0)
```

```python
X = np.random.uniform(-0.1,0.1,(200,2))
X[:100,1] += 1
X[:,0] *= 10
y = X[:,1]>0.5
X[:100,1] += (X[:100,0]+1)**2
X = X @ [[1,1],[1,0]]
X.shape, y.shape
```

```python
import matplotlib.pyplot as plt
for k in set(y):
    plt.plot(X[:,0][y==k], X[:,1][y==k], 'o')
plt.show()
```

```python
Xpca = pca.fit_transform(X)
cor = ['r', 'b']
for k in set(y):
    plt.plot(Xpca[:,0][y==k], np.zeros((sum(y==k),)), f"o{cor[int(k)]}")
    plt.plot(Xpca[:,0][y==k], Xpca[:,1][y==k], f"o{cor[int(k)]}")
plt.show()
```

```python
np.var(X, axis=0), np.var(Xpca, axis=0)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
Xs = scaler.fit_transform(X)

for k in set(y):
    plt.plot(Xs[:,0][y==k], Xs[:,1][y==k], 'o')
plt.show()
cor = ['r', 'b']
Xpca = pca.fit_transform(X)
for k in set(y):
    plt.plot(Xpca[:,0][y==k], np.zeros((sum(y==k),)), f"o{cor[int(k)]}")
    plt.plot(Xpca[:,0][y==k], Xpca[:,1][y==k], f"o{cor[int(k)]}")
plt.show()
np.var(X, axis=0), np.var(Xpca, axis=0)
```

```python
from sklearn.decomposition import KernelPCA
kpca = KernelPCA(kernel='poly',degree=2)

for k in set(y):
    plt.plot(X[:,0][y==k], X[:,1][y==k], 'o')
plt.show()
cor = ['C0', 'C1']
Xpca = kpca.fit_transform(X)
for k in set(y):
    plt.plot(Xpca[:,0][y==k], np.zeros((sum(y==k),)), f"o{cor[int(k)]}")
    plt.plot(Xpca[:,0][y==k], Xpca[:,1][y==k], f"o{cor[int(k)]}")
plt.show()
np.var(X, axis=0), np.var(Xpca, axis=0)
```

```python
from sklearn.manifold import TSNE

X, y = iris.data, iris.target

for k in set(y):
    plt.plot(X[:,0][y==k], X[:,1][y==k], 'o')
plt.show()
cor = ['C0', 'C1', 'C2']
X2 = TSNE(init='random').fit_transform(X)
for k in set(y):
    plt.plot(X2[:,0][y==k], X2[:,1][y==k], f"o{cor[int(k)]}")
plt.show()
np.var(X, axis=0), np.var(X2, axis=0)
```
